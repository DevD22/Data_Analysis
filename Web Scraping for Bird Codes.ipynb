#### Importing Packages

```python
import requests
from bs4 import BeautifulSoup
```

#### Making a GET request
Website scraped for Bird Codes:
https://www.birdsontario.org/species-codes/

```python
base_site="https://www.birdsontario.org/species-codes/"
response=requests.get(base_site)
response  # 200 indiciates a succeful get request
```

```python
# Checking the response material
html=response.content
html[:100]
```

```python
# Since we have scarpped the website, we need t make use of the HTML code suing a parser
soup=BeautifulSoup(html,"html.parser")
```

```python
# Exporting HTML to a file
with open("BirdCodes.html",'wb')as file:
    file.write(soup.prettify('utf-8'))
```

#### Navigating and Searching the HTML tree

```python
# exploring .find() and .find_all()
#.find() returns the first tag matching the search
#.find_all() returns a list of all tags
links=soup.find_all('tbody') 
links
```

```python
k=soup.find_all('td')# Length 1024
type(k)
```

```python
import pandas as pd
df=pd.DataFrame(columns=['Code','Name'])

for ind,text in enumerate(k):
    if len(text.get_text())<=4:
        df.loc[ind,'Code']=text.get_text()
    else:
        df.loc[ind,'Name']=text.get_text()
    
#     if count%2==0:
#         df['Code']=i.get_text()
#     else:
#         df['Name']=i.get_text()
#     count+=1
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(df)
```

```python
df2=df.copy()
```

```python
for ind in df2.columns:
    df2[ind] = df2[ind].replace('NaN', pd.np.nan)

for column in df2.columns:
    df2[column].dropna(inplace=True) 
```

```python
code=[]
for column in df2['Code']:
    if type(column)==str:
        code.append(column)

name=[]

for column in df2['Name']:
    if type(column)==str:
        name.append(column)
name
```

```python
new_code = list(filter(lambda string: string != "", code))
new_name = list(filter(lambda string: string != "", name))
```
